# Munic√≠pio da Cidade de Lichinga

```python
# Importar bibliotecas
import osmnx as ox
import matplotlib.pyplot as plt
from IPython.display import display

# Obter o pol√≠gono da prov√≠ncia do Niassa
niassa = ox.geocode_to_gdf("Niassa, Mozambique")

# Definir os filtros para munic√≠pios (admin_level=8)
tags = {
    "boundary": "administrative",
    "admin_level": "8"
}

# Obter todos os munic√≠pios dentro do Niassa
municipios_niassa = ox.features_from_polygon(niassa.geometry.iloc[0], tags)

# Filtrar especificamente o munic√≠pio da cidade de Lichinga
cidade_lichinga = municipios_niassa[municipios_niassa["name"].str.contains("Cidade de Lichinga", case=False, na=False)]

# Se n√£o encontrar com esse nome exato, tentar por "Lichinga" apenas
if cidade_lichinga.empty:
    cidade_lichinga = municipios_niassa[municipios_niassa["name"].str.contains("Lichinga", case=False, na=False)]

# Visualizar os dados
display(cidade_lichinga[["name", "geometry"]])

# Plotar
cidade_lichinga.plot(edgecolor='black', facecolor='orange', figsize=(8, 8))
plt.title("Munic√≠pio da Cidade de Lichinga")
plt.axis('off')
plt.show()
```

```python
# Baixar rede rodovi√°ria
from networkx import edges
import osmnx as ox

L = ox.graph_from_polygon(cidade_lichinga.geometry.to_list()[0])

print('N√∫mero de intersec√ß√µes: ', L.number_of_nodes())
print('N√∫mero de segmentos de estrada: ', L.number_of_edges())

# Definindo o gr√°fico dentro do GeoDataFrame
nodes, edges = ox.graph_to_gdfs(L)
```

```python
# Plotando os pontos do GeoDataFrame e extraindo da rede rodovi√°ria
from turtle import color
import matplotlib.pyplot as plt
f, ax = plt.subplots(1, 2, figsize=(16, 8))

nodes.plot(ax=ax[0], color = 'k', markersize = 5)
edges.plot(ax=ax[1], color = 'k', linewidth = 1)

plt.tight_layout()
```

```python
# Criar um rascunho de visualiza√ß√£o
f, ax = plt.subplots(figsize = (16, 8))

edges.plot(ax=ax, linewidth =0.5, color = 'darkblue')
ax.set_title("Munic√≠pio da Cidade de Lichinga")

plt.tight_layout()
plt.show()
```

```python
# Defini√ß√£o de cores
from matplotlib import colors
import matplotlib.pyplot as plt
from seaborn import color_palette

# Selecionando paleta de cores
# color_palette = [{"Neon": "#ff0099",  "Background": "#2e002e"}]
neon = "#ff0099"
background = "#2e002e"


# Criar Gr√°fico
f, ax = plt.subplots(1,1, figsize=(12,12))
edges.plot(ax=ax, color = neon, linewidth = 0.5, alpha = 0.9)
ax.set_facecolor(background)

ax.set_xticks([])
ax.set_yticks([])
for spine in ax.spines.values():
    spine.set_visible(False)

# Adicionar o t√≠tulo com espa√ßo extra no eixo Y
ymin, ymax = plt.ylim()
extension = 0.1 * (ymax - ymin)
ax.set_ylim(ymin, ymax + extension)
ax.set_title("Cidade de Lichinga", fontsize=20, color=neon, y=0.9)

plt.show()
```
